Далее выгрузим информацию по каждому датасету и проведем предварительное исследование. 

```python
## функция, которая выводила бы сразу сведенную информацию в рамках исследования и предобработки.

def lookup_datasets(dfs, filenames=None):
    """
    Функция для вывода общей информации о датасетах из списка.

    Параметры:
        dfs (list of pd.DataFrame): Список датафреймов для обработки.
        filenames (list of str): Список названий файлов (имена для отображения).
    """
    if filenames is None:
        filenames = [f"df_{i + 1}" for i in range(len(dfs))]

    for df, df_name in zip(dfs, filenames):
        print('-'*40, df_name, '-'*40)
        
        # Пропущенные значения
        missing_data = df.isna().mean()
        missing_data_result = missing_data.apply(lambda x: f'{x:.2%}' if x > 0 else "")
        missing_data_name = "Пропущено" if missing_data.sum() > 0 else ""
        missing_data_result.name = missing_data_name

        # Типы данных
        dtypes_result = df.dtypes
        dtypes_result.name = "Тип данных колонки"
        fewest_nans_row = df.iloc[1:-1].isna().sum(axis=1).idxmin()

        values_type = df.loc[fewest_nans_row].map(type).T
        values_type.name = "Тип значения"
        
        ## Вывод первой строки
        first_row = df.iloc[0,:]
        first_row.name = "Первая строка"
        
        # вывод строки без пропусков
        full_row = df.loc[fewest_nans_row].T
        full_row.name = "Полная строка"
        
        #вывод  последней строки
        finish_row = df.iloc[-1, :]
        finish_row.name = "Последняя строка"
        
        # Объединяем результаты и пример данных
        result = pd.concat([
            dtypes_result, # типы
            values_type, # типы значений
            missing_data_result, # пропущенные значения
            first_row, # первая строка
            full_row, # полная строка
            finish_row # последняя строка
        ], axis=1)

        display(result)

        # Проверка на полные дубликаты
        duplicates = df.duplicated().mean()
        if duplicates > 0:
            print(f'Полных дубликатов: {duplicates:.2%}')
```

### 3.3. Анализ аномальных продаж

Исследуем случаи, когда компании были куплены за нулевую сумму или символический доллар ($1), несмотря на то, что они получили значительные вложения (ненулевое значение в столбце `funding_total`). Эти ситуации могут вызывать сомнения в достоверности данных.


```python
#фильтрация данных 
condition_purchased = ((startup_full['price_amount']==0)|(startup_full['price_amount']==1))&(startup_full['funding_total']>0) 
free_purchased = startup_full[condition_purchased]
free_purchased.reset_index(drop=True, inplace=True)
```

```python
free_purchased.head(15)
```

```python
free_purchased.info()
```

Как мы видим в `free_purchased` представлено достаточно много компаний, которые были приобретены за символический 0 или 1 долларов. Продажа компании по символической цене может быть совокупностью множества факторов, в том числе:
- результат банкротства, то есть новый собственник фактически приобретёт компанию за 1 доллар, но при этом новому владельцу переходят все обязательства компании;
-  передача компании в управлении новым инвесторам;
-  Стратегические мотивы: Продажа за символическую сумму может быть частью стратегии слияния или поглощения, когда одна компания приобретает другую для расширения своего портфеля или доступа к технологиям;
- Регуляторные ограничения: В некоторых случаях продажа за символическую сумму может быть обусловлена требованиями регулирующих органов или условиями соглашений с кредиторами.

```python
free_purchased['funding_total'].describe()
```

Уже в данных полученных в describe() мы видим достаточно большой размах между 25% и 75%, о чем ранее мы говорили, поэтому, проведём расчет границ выбросов исходя из данных.

```python
q1 = free_purchased['funding_total'].quantile(0.25)
q3 = free_purchased['funding_total'].quantile(0.75)
iqr = q3 - q1

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
print(f'Нижняя граница выбросов: {lower_bound}')
print(f'Верхняя граница выбросов: {upper_bound}')
```

Отрицательное значение нижней границы говорит нам о наличии отрицательных значений в рассчитанных данных говорит нам о наличии аномальных или некорректных данных в выборке, что ранее мы также упоминали при исследовании данных в п.3.2. В данном случай значение отрицательным не может быть. так как минимальная сумму финансирования проекта - это 0, но никак не  - `$18 250 000`. К том уже стоит отметить, что представленная ранее выборка у нас ассиметрична, когда основная часть данных собрана в диапазоне до `$100 млн`.

```python
# Исправленная нижняя граница (чтобы она была неотрицательной)
lower_bound_cor = max(q1 - 1.5 * iqr, 0)
```

```python
# Количество значений ниже нижней границы
values_below_lower_bound = (free_purchased['funding_total'] <= lower_bound_cor).sum()

# Количество значений выше верхней границы
values_above_upper_bound = (free_purchased['funding_total'] >= upper_bound).sum()

# Общее количество наблюдений
total_values = len(free_purchased)

# Процентили для нижней и верхней границ
lower_percentile = values_below_lower_bound / total_values * 100
upper_percentile = values_above_upper_bound / total_values * 100

print(f'Нижний процентиль: {lower_percentile:.2f}%')
print(f'Верхний процентиль: {upper_percentile:.2f}%')
```

Замена нижней границы выбросов на значение, соответствующее нулевому процентилю, не приведет к изменениям, так как все значения и так находятся выше этой границы. Замена верхней границы выбросов на значение, соответствующее 91-му процентилю (100% - 9.09% ≈ 91%), позволит исключить выбросы и сосредоточиться на основном диапазоне данных. Это может быть целесообразным шагом, если цель заключается в улучшении качества анализа путем исключения экстремальных значений.

### 3.5. Оценка раундов финансирования стартапа перед покупкой

Необходимо проанализировать столбец `funding_rounds`Выяснить типичное количество раундов финансирования для стартапов, находящихся в различных состояниях (status). Мы исследуем зависимость между статусом стартапа и количеством прошедших раундов финансирования.


Для исследования раундов  исследуем данные только по тем компаниям, которы  получили финансирование.

```python
print(startup_full['funding_rounds'].describe())
```

```python
grouped_rounds = startup_full.groupby('status')['funding_rounds'].median()
print(grouped_rounds)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(grouped_rounds.index, grouped_rounds.values)
plt.xlabel('Статус')
plt.ylabel('Типовое количество раундов финансирования ')
plt.title('Типовое количество раундов финансирования в зависимости от статуса')
plt.grid(False)
plt.show()
```

График показывает, что стартапы, достигшие стадии продажи или IPO, проходили в среднем два раунда финансирования, тогда как закрытые стартапы и стартапы на этапе открытия прошли всего один раунд. 

- Успешные стартапы – это проекты, которые были проданы или же вышли на IPO в среднем получают не менее 2 раундов финансирования, что говорит нам о том, что для достижения успеха компании требуется получить значительно число средств для финансирования текущей деятельности. Однако хорошим гарантом   получения данных средств является сильные бизнес-модели, устойчивый рост и высокий потенциал, что привлекает инвесторов на протяжении нескольких этапов.
- Неудачные стартапы (закрытые) проходят в среднем один раунд финансирования. Это может быть результатом того, что компании столкнулись с проблемами в самом начале функционирования, в том числе снижение спроса, проблемы управления, высокая конкуренция и так далее. После первого раунда они могли не показать достаточного прогресса, чтобы привлечь следующий раунд инвестиций.
- Стартапы на этапе открытия также получают в среднем 1  раунд финансирования, что является достаточно логичным результатам, так как они находятся только на начальной стадии реализации идеи и бизнес – стратегии. Первый раунд финансирования может использоваться для тестирования идеи и проверки её жизнеспособности.

Можно сделать общий вывод о том, что успешные стартапы требуют больше количества раундов финансирования для поддержания и обеспечения стабильного роста и им лучше им его получить, а нежизнеспособные проекты могут быть закрыты уде после первого этапа финансирования.
